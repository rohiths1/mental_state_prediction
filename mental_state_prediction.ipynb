```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
print(train.shape)
print(test.shape)
train.head()
```

```python
test.head()
```

```python
train = train[["ID", "job", "country", "state"]]
test = test[["ID", "job", "country"]]
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
train.dropna()
test.dropna()
```

```python
train["state_num"] = train["state"].map({'Worthlessness':0,'Anxiety':1,'Death':2,'Despair':3,'Dread':4,'Fatigue':5,'Nausea':6,'Pain':7,'Stress':8,'Tears':9})
train.groupby('state').size()
```

```python
train = train.drop('state', axis = 1)
train = train.drop('ID', axis = 1)
train.head()
```

```python
X_train = train.drop('state_num',axis = 1)
y_train = train['state_num']
X_train = pd.get_dummies(X_train)
test = pd.get_dummies(test)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
y_pred = logreg.predict(test.drop('ID', axis=1))
y_pred
```

```python
table = np.array([test["ID"], y_pred])
table = np.transpose(table)
table
```

```python
submission = pd.DataFrame(table)
submission
```

```python
submission.columns = ["ID", "state_num"]
submission
```

```python
submission["state"] = submission["state_num"].map({0:'Worthlessness',1:'Anxiety',2:'Death',3:'Despair',4:'Dread',5:'Fatigue',6:'Nausea',7:'Pain',8:'Stress',9:'Tears'})
submission = submission.drop("state_num", axis = 1)
submission
```

```python
submission.to_csv("solution.csv", index = False)
```
